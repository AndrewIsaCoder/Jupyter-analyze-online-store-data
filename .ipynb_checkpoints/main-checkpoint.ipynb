## 3. üíæ PregƒÉtirea datelor

√én aceastƒÉ etapƒÉ √ÆncƒÉrcƒÉm datele din fi»ôierul CSV care con»õine informa»õii despre produsele disponibile √Æn magazinul online.  

Vom realiza o privire de ansamblu asupra structurii datasetului:  
- Vom afi»ôa primele r√¢nduri;
- Vom verifica tipurile de coloane »ôi dacƒÉ existƒÉ valori lipsƒÉ;
- Vom pregƒÉti eventuale coloane noi necesare pentru analizƒÉ.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("PDV5_02-Task_2-online_store_data.csv")

df.head() # Afisarea primelor 5 randuri
```

## 4. üéØ PerspectivƒÉ analiticƒÉ 1 ‚Äì Valorile extreme ale evaluƒÉrilor

Pentru a √Æn»õelege cum percep clien»õii produsele, este esen»õial sƒÉ identificƒÉm nota minimƒÉ »ôi maximƒÉ acordatƒÉ de utilizatori.  
Aceste valori ne oferƒÉ o imagine de ansamblu asupra extremelor din experien»õa clien»õilor.

Mai jos extragem cele douƒÉ valori »ôi le prezentƒÉm textual:

```python
print("üìà Nota maxima : " , df["rating"].max())
print("üìâ Nota minima : " , df["rating"].min())
```

## 5. üìà PerspectivƒÉ analiticƒÉ 2 ‚Äì Evolu»õia evaluƒÉrilor √Æn timp

Pentru a analiza dacƒÉ percep»õia clien»õilor asupra produselor s-a schimbat √Æn timp, vom examina evolu»õia notelor medii pe ani.

AceastƒÉ tendin»õƒÉ ne poate indica:
- dacƒÉ produsele mai recente sunt mai bine sau mai slab primite;
- cum se schimbƒÉ satisfac»õia clien»õilor √Æn func»õie de perioada √Æn care un produs este introdus √Æn ofertƒÉ.


```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
evolutie = df.groupby('date_added')['rating'].mean().reset_index()

plt.figure(figsize=(12 , 6))
sns.lineplot(data = evolutie , x = "date_added" , y = "rating" , marker="o" , color="blue")

plt.title("Evolu»õia notelor medii √Æn func»õie de anul adƒÉugƒÉrii produselor")
plt.xlabel("Anul adaugarii produselor: ")
plt.ylabel("Nota medie: ")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 6. üóÇÔ∏è PerspectivƒÉ analiticƒÉ 3 ‚Äì Diferen»õele dintre note √Æn func»õie de categorii

Pentru a √Æn»õelege ce tipuri de produse sunt cele mai apreciate de clien»õi, analizƒÉm **nota medie pe fiecare categorie** de produse.

AceastƒÉ analizƒÉ ne ajutƒÉ sƒÉ rƒÉspundem la √ÆntrebƒÉri precum:
- Ce tipuri de produse au avut cel mai mare succes?
- ExistƒÉ categorii unde satisfac»õia clien»õilor este sub medie?


```python
categorii = df.groupby("category")["rating"].mean().reset_index()
categorii = categorii.sort_values(by="rating" , ascending=False)

plt.figure(figsize=(14,6))
sns.barplot(data=categorii , x = "category" , y = "rating" , palette = "viridis")

plt.title("Rating mediu pe categorii: ")
plt.xlabel("Categorii: ")
plt.ylabel("Rating mediu: ")
plt.tight_layout()
plt.show()
```

## 7. üéØ Reprezentarea vizualƒÉ a rezultatelor

Pentru a sprijini concluziile ob»õinute √Æn pa»ôii anteriori, am inclus urmƒÉtoarele grafice care eviden»õiazƒÉ vizual tendin»õele »ôi diferen»õele observate √Æn date:


‚úÖ Grafic 1: Evolu»õia notelor medii √Æn timp (linia de tendin»õƒÉ)

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=evolutie, x='date_added', y='rating', marker='o', color='steelblue')
plt.title("Evolu»õia notelor medii √Æn func»õie de anul adƒÉugƒÉrii")
plt.xlabel("Anul adƒÉugƒÉrii produselor")
plt.ylabel("NotƒÉ medie")
plt.grid(True)
plt.tight_layout()
plt.show()
```

üîé **Interpretare**:
Graficul aratƒÉ cum s-a modificat nota medie acordatƒÉ produselor de-a lungul timpului. ObservƒÉm (ex. cre»ôtere/declin/stabilitate) √Æn anumite perioade, suger√¢nd posibile schimbƒÉri √Æn calitatea produselor sau √Æn comportamentul utilizatorilor.


‚úÖ Grafic 2: Note medii pe categorii (bar chart orizontal)

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=categorii, x='rating', y='category', palette='muted')
plt.title("Nota medie a produselor pe categorii")
plt.xlabel("NotƒÉ medie")
plt.ylabel("Categorie")
plt.tight_layout()
plt.show()
```

üîç **Interpretare**:
Acest grafic ne permite sƒÉ comparƒÉm rapid performan»õa categoriilor. Categoriile din partea de sus au cele mai bune evaluƒÉri medii »ôi pot indica produse de succes sau bine pozi»õionate. Categoriile din partea de jos ar putea necesita √ÆmbunƒÉtƒÉ»õiri sau investiga»õii suplimentare.
